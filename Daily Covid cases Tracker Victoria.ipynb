```python
# Creating a function to automate the price check process every day

def covid_tracker():
    URL = 'https://covidlive.com.au/vic'

    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    page = requests.get(URL, headers=headers)

    soup1 = BeautifulSoup(page.content, "html.parser")

    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
    
    table = soup.find_all('table')[1]
    covid_headers = table.find_all('th')
    covid_table_header = [header.text.strip() for header in covid_headers]
    
    import pandas as pd
    df1 = pd.DataFrame(columns = covid_table_header)
    column_data = table.find_all('tr')
    
    # Getting the data for the columns as the data is stored between <td> tags
    for row in column_data[1:]: # since the first list is empty
        row_data = row.find_all('td')
        indivi_row_data = [data.text.strip() for data in row_data]
        #print(indivi_row_data)
        length = len(df1)  
        df1.loc[length] = indivi_row_data

    total = df1['TOTAL'].loc[df1.index[0]]

    seven_day_avg = df1['TOTAL'].loc[df1.index[1]]
    
    import datetime

    today = datetime.date.today()
    
    import csv 

    headers_csv = ['Total_Cases','7_Day_Avg','Date']
    data = [total,seven_day_avg,today]

    with open('CovidTracker.csv', 'a+', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)
    
    if(seven_day_avg > 200):
      send_mail()
```
