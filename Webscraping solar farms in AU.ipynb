```python
# Webscraping using Python to find out the list of largest solar farms in Australia
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Solar_power_in_Australia'
```

```python
page = requests.get(url)
```

```python
print(page)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[6]
```

```python
# Finding the 6th table based on the webpage tables
table = soup.find_all('table')[6]
print(table)
```

```python
# Getting the headers of the table from the webpage from 6th table
solar_headers = table.find_all('th')
```

```python
print(solar_headers)
```

```python
solar_table_header = [header.text.strip() for header in solar_headers]
print(solar_table_header)
```

```python
import pandas as pd
```

```python
# Creating the column headers for the dataframe
df = pd.DataFrame(columns = solar_table_header)
df
```

```python
column_data = table.find_all('tr')
```

```python
print(column_data)
```

```python
# Getting the data for the columns as the data is stored between <td> tags
for row in column_data[1:]: # since the first list is empty
    row_data = row.find_all('td')
    indivi_row_data = [data.text.strip() for data in row_data]
    #print(indivi_row_data)
    length = len(df)  
    df.loc[length] = indivi_row_data
```
